````python
import threading

in_numbers = [[33, 55, 21, 55], [87, 43, 62, 34, 66, 11], [86, 41, 52, 67, 84, 11, 2], [1, 3, 5]]
my_sum = 0
lock = threading.Lock()

## BEGIN SOLUTION
def work_thread(elem):
    global my_sum
    with lock:
        my_sum += sum(elem)
        

threads = []
for i in range(len(in_numbers)):
    t = threading.Thread(target=work_thread, args=(in_numbers[i],))
    t.start()
    threads.append(t)
for t in threads:
    t.join()

print(my_sum)
### END SOLUTION
````
